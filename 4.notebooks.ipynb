## 1. OVERVIEW AND IMBALANCE ANALYSIS

### Extreme Imbalance:
* **Evidence:** The DELUX customer segment accounts for **99.61%** of the total customer file, while the VIP segment represents only **0.04%** (2 customers).
* **Insight:** The business model operates almost entirely on a low-value/high-churn risk customer base. The revenue risk is **extreme** if the 2 VIP customers cease transactions.

### Dormancy Risk:
* **Evidence:** The average Recency (R) for DELUX customers is **201 days**.
* **Insight:** Most customers have not made a purchase in over 6.5 months. This signals a **customer retention crisis** and necessitates urgent "Win-back" campaigns.

## 2. DEEP DIVE SEGMENT ANALYSIS

### 2.1. VIP Segment (0.04% - Critically Important)
* **Evidence (F):** VIP customers have an average purchase Frequency (F) of **344 times**, which is **63.7 times higher** than the DELUX group (5.4 times).
* **Evidence (M):** The average Monetary Value (M) is **$115,895.60**, which is **79 times higher** than DELUX ($1,465.39).
* **Insight (Action):** This group is the **Lifeblood** of the business. All strategies must focus on maintaining maximum Recency ($\text{R}=1$ day) through **Personalized Premium Service** to ensure zero friction in their purchasing journey.

### 2.2. STANDARD Segment (0.35% - Potential Pool)
* **Evidence (R):** This segment exhibits good Recency (R) at **24 days**.
* **Evidence (F):** The purchase Frequency (F) is **109 times**, which is **20 times higher** than the DELUX group (5.4 times).
* **Insight (Action):** This is the **Growth Pool** and should be treated as the next generation of VIPs. Implement **Upsell/Cross-sell** strategies via high-value product bundles, aiming to increase their average Monetary Value ($\text{M}$) above the $50,000$ threshold to transition this group to VIP status.

### 2.3. DELUX Segment (99.61% - Risk & Cost)
* **Evidence (R & F):** This group has a poor average Recency (R) of **201 days**, with low purchase Frequency (F) at only 5.4 times.
* **Evidence (M):** Average spending is low, at **$1,465.39**.
* **Insight (Action):** Using mass marketing budgets on this group is **wasteful**. Focus on **Re-segmentation** within the DELUX group based on their last known Recency. Run **Low-Cost Nudges** (e.g., reminder emails, free shipping offers) to activate a small fraction of the segment, accepting that the majority may be permanently lost.
